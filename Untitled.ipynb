```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
import sys
```

```python
# Cambiar el directorio de trabajo
os.chdir('/home/wirakocha/Documentos/GITHUB/HOBO_temp/data/')

# Obtener la lista de archivos CSV
myFiles = glob.glob('*.csv')
print(myFiles)
myFiles[1]
```

```python
data = pd.DataFrame(columns=['count', 'tem'])
date = []

for i in range(len(myFiles)):
    # Cargar el archivo .csv
    file = pd.read_csv(myFiles[i])
    
    # Determinar si existe 'Temp' y en qué posición
    tdx = [k for k, col in enumerate(file.columns) if 'Temp' in col]
    for k in tdx:
        if len(tdx) > 0:
            file.columns.values[k] = 'tem'
            tem = file['tem']
            
            # Convertir valores negativos en ceros
            tem = tem.apply(lambda x: x if x >= 0 else 0)
            tem = tem.apply(lambda x: x if x <= 50 else 0)
            
            # Estandarizar los valores missing
            tem = tem.replace(0, pd.NA)
            
            # Rellenar los datos missing
            tem = tem.interpolate(method='linear')
            
    # Determinar si existe 'Fecha' y en qué posición
    fdx = [k for k, col in enumerate(file.columns) if 'Fecha' in col]
    for k in fdx:
        if len(fdx) > 0:
            file.columns.values[k] = 'fecha'
            fecha = file['fecha']
    
    count = file['N__']
    
    # Eliminar datos iniciales y finales asociados a inmersión y emersión del sensor
    out = 2
    plt.plot(fecha, tem, '-ok')
    plt.hold(True)
    g = np.gradient(tem)
    idx = np.where(g == 0)[0]
    plt.plot(fecha[idx[out]:idx[-out]],
             tem[idx[out]:idx[-out]], 'or')
    tem = tem[idx[out]:idx[-out]]
    count = count[idx[out]:idx[-out]]
    fecha = fecha[idx[out]:idx[-out]]
    
    # Unir las matrices de distintas fechas
    date.extend(fecha)
    data = pd.concat([data, pd.DataFrame({'count': count, 'tem': tem})])

    # Plot variables en el tiempo
    plt.figure(1)
    plt.show()

data = pd.DataFrame({'count': data['count'], 'tem': data['tem']})
data = data.set_index(pd.DatetimeIndex(date))
```
